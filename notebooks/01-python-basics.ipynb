## 2. Variables and Data Types

Variables are containers that store values. Python has different data types:

```python
# Strings (text)
name = "Alice"
country = "Nigeria"
print(f"Hello, I'm {name} from {country}")
```

```python
# Numbers - Integers
age = 25
year_born = 2001
print(f"I am {age} years old")
print(f"I was born in {year_born}")
```

```python
# Numbers - Floats (decimals)
height = 5.9
weight = 65.5
print(f"Height: {height} ft")
print(f"Weight: {weight} kg")
```

```python
# Booleans (True or False)
is_student = True
is_working = False
print(f"Student: {is_student}")
print(f"Working: {is_working}")
```

## 3. Basic Arithmetic Operations

```python
# Addition and Subtraction
a = 10
b = 3

print(f"{a} + {b} = {a + b}")
print(f"{a} - {b} = {a - b}")
```

```python
# Multiplication and Division
print(f"{a} * {b} = {a * b}")
print(f"{a} / {b} = {a / b}")
```

```python
# Power and Modulo (remainder)
print(f"{a} ** {b} = {a ** b}")  # 10 to the power of 3
print(f"{a} % {b} = {a % b}")    # Remainder when 10 is divided by 3
```

## 4. String Operations

```python
# String concatenation
first_name = "John"
last_name = "Doe"
full_name = first_name + " " + last_name
print(full_name)
```

```python
# String repetition
word = "Python"
print(word * 3)  # Repeats the string 3 times
```

```python
# String length
message = "Hello, Data Analysis!"
print(f"Message: {message}")
print(f"Length: {len(message)} characters")
```

## 5. Comparisons and Boolean Operations

```python
# Comparisons return True or False
x = 10
y = 5

print(f"{x} > {y}: {x > y}")
print(f"{x} < {y}: {x < y}")
print(f"{x} == {y}: {x == y}")
print(f"{x} != {y}: {x != y}")
```

```python
# Boolean operations: and, or, not
is_sunny = True
is_warm = True

print(f"Sunny AND Warm: {is_sunny and is_warm}")
print(f"Sunny OR Warm: {is_sunny or is_warm}")
print(f"NOT Sunny: {not is_sunny}")
```

## 6. If/Else Statements

```python
# Simple if/else
temperature = 25

if temperature > 30:
    print("It's hot!")
elif temperature > 20:
    print("It's warm")
else:
    print("It's cold")
```

```python
# Another example: Grade checker
score = 85

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
else:
    grade = "F"

print(f"Score: {score}")
print(f"Grade: {grade}")
```

## 7. Loops - For Loop

```python
# For loop with range
print("Counting to 5:")
for i in range(5):
    print(i)
```

```python
# For loop with range(start, stop, step)
print("Counting by 2s from 0 to 10:")
for i in range(0, 10, 2):
    print(i)
```

```python
# For loop over a list
fruits = ["apple", "banana", "cherry"]
print("Fruits:")
for fruit in fruits:
    print(f"- {fruit}")
```

## 8. Loops - While Loop

```python
# While loop
count = 0
while count < 5:
    print(f"Count: {count}")
    count += 1  # Same as: count = count + 1
print("Done!")
```
